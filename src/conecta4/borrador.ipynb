```python
def repr(tablero)->str:
    """
    Pregunta a grokk que hacer __repr__ y sobreescribe el metodo
    """
    tablero_impreso = ""
    for col in tablero:
        tablero_impreso += str(col) + "\n"
    return tablero_impreso
```

```python
print(repr(columns))
```
